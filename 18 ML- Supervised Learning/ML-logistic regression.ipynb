```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

data = {
    "Glucose": [80, 90, 100, 120, 140, 160, 180, 200],
    "BMI": [22, 24, 25, 28, 30, 32, 35, 38],
    "Diabetes": [0, 0, 0, 0, 1, 1, 1, 1]
}

df = pd.DataFrame(data)

X = df[["Glucose", "BMI"]]
y = df["Diabetes"]

model = LogisticRegression()
model.fit(X, y)

Glucose = int(input("Enter Glucose Level: "))
BMI = float(input("Enter BMI: "))
new_data = pd.DataFrame({
    'Glucose': [Glucose],
    'BMI': [BMI]})

prediction = model.predict(new_data)

if prediction[0] == 1:
    print("Diabetes Detected")
else:
    print("No Diabetes")
    
```
