```python
from nltk.tokenize import sent_tokenize
import nltk
nltk.download('punkt_tab')
data ='Every object in the universe continues to be in the state of rest or uniform motion unless an external force is applies on it.'
print(sent_tokenize(data))
```

```python
from nltk.tokenize import word_tokenize
data1 = 'Every object in the universe continues to be in the state of rest or uniform motion unless an external force is applies on it.'
print(word_tokenize(data1))
```
