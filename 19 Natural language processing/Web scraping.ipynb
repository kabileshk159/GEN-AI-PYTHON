```python
# to fetch a content from url
import urllib.request
data= urllib.request.urlopen("https://www.linkedin.com")
content= data.read()

# to fetch only the text from the html content
from bs4 import BeautifulSoup
soup=BeautifulSoup(content)
content_data= soup.get_text(strip=True)
print(content_data)

# to separate words and see
words= [i for i in content_data.split()]
print(words)

#to tell repition of each words
import nltk
fq = nltk.FreqDist(words) 
for k,v in fq.items():
    print(str(k),':',str(v))

#visualizing the freq dist
fq.plot(5)
```
