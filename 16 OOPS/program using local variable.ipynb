```python
class Employee:

    def get_details(self):
        emp_name = "Kabilesh"
        emp_id = 101            
        emp_role = "AI Engineer" 

        print("Employee Details")
        print("----------------")
        print("Name:", emp_name)
        print("ID:", emp_id)
        print("Role:", emp_role)

    def calculate_salary(self):
        basic_salary = 30000     
        bonus = 5000            
        total_salary = basic_salary + bonus

        print("Salary Details")
        print("----------------")
        print("Basic Salary:", basic_salary)
        print("Bonus:", bonus)
        print("Total Salary:", total_salary)


employee1 = Employee()

employee1.get_details()
employee1.calculate_salary()
```
