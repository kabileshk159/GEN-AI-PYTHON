```python
from abc import ABC, abstractmethod

# Abstract class
class LoginSystem(ABC):
    def login(self, username, password):
        pass


# Child class
class Website(LoginSystem):

    def login(self, username, password):
        if username == "admin" and password == "1234":
            print("Login Successful")
        else:
            print("Invalid Login")


# Object
user = Website()
user.login("admin", "1234")
user.login("user", "124")
```
