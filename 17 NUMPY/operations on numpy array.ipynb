```python
import numpy as np

# Creating NumPy arrays
numbers = np.array([10, 20, 30, 40, 50])

print("Original Array:")
print(numbers)

# Accessing elements
print("First element:")
print(numbers[0])

print("Last element:")
print(numbers[-1])

# Array operations
print("Addition with 5:")
print(numbers + 5)

print("Multiplication by 2:")
print(numbers * 2)

# Finding properties
print("Array size:")
print(numbers.size)

print("Array datatype:")
print(numbers.dtype)

```
