```python
# to print a word from the numpy array
import numpy as np

arr = np.array([
    [
        ['H', 'E', 'L'],
        ['A', 'B', 'C']
    ],
    [
        ['L', 'O', 'X'],
        ['Y', 'Z', 'D']
    ]
])

word = (
    arr[0, 0, 0] +  
    arr[0, 0, 1] +  
    arr[0, 0, 2] +  
    arr[1, 0, 0] +  
    arr[1, 0, 1]    )

print(word)
```
